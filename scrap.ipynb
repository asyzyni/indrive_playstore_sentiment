```python
import pandas as pd
from google_play_scraper import Sort, reviews_all

# Ambil semua atau tentukan target review (misal com.twitter.android atau app id pilihan Anda)
result = reviews_all(
    'sinet.startup.inDriver', # Ganti dengan App ID yang ingin Anda scrape
    lang='id',             # Filter Bahasa Indonesia
    country='id',          # Regional Indonesia
    sort=Sort.MOST_RELEVANT, # Menggunakan Most Relevant agar variasi datanya bagus
    filter_score_with=None # Mengambil semua rating (1 sampai 5)
)

# Konversi hasil ke DataFrame Pandas
df = pd.DataFrame(result)

# Filter data berdasarkan tahun (Minimal Tahun 2024 sesuai aturan tugas)
df['at'] = pd.to_datetime(df['at'])
df_filtered = df[df['at'].dt.year >= 2026]

# Cetak jumlah data yang berhasil didapatkan
print(f"Total data review terkumpul: {len(df_filtered)}")

# Simpan ke CSV untuk proses Text Mining berikutnya
df_filtered.to_csv('data/reviews_playstore_indonesia.csv', index=False, encoding='utf-8')
```
